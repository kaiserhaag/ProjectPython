```python
#import dependencies

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#file to load
temperature_data = "temperature.csv"
crop_data = "crop_production.csv"

#read the temperature and crop data, encoding='ISO-8859-1'
temp_data = pd.read_csv(temperature_data,encoding='ISO-8859-1')
crops_data = pd.read_csv(crop_data,encoding='ISO-8859-1')


#before merge we need to change the column name on crop data [area to country]
crops_clean= crops_data.rename(columns={"Area": "Country"})
crops_clean.head()
```

```python
#clean temperature dataset 
temp_data.head()
temp_clean= temp_data.groupby(['Country',''()
temp_clean.head()
```

```python
#combine into a single dataset
temp_crop_data= pd.merge(crops_clean,temp_data, how= "left", on=["Country"])

#display the table 
temp_crop_data.head()
```
