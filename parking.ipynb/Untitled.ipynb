```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# file path
temp = "../temperature_clean.csv"
production = "../crop_production_clean.csv"
```

```python
# read csv files
temp_df = pd.read_csv(temp)
#temp_df.head()
```

```python

crop_pro_results = pd.read_csv(production, encoding="ISO-8859-1")
crop_pro_results.head()
```

```python
rice_df = crop_pro_results[crop_pro_results['Item']=="Rice (Milled Equivalent)"]
rice_top_df = rice_df.nlargest(10, "Y1961")
rice_top_df['rice production'] = rice_top_df.sum(axis=1)
rice_top_df.head(2)
final_rice_df = rice_top_df.sort_values(by=['rice production'], ascending=False)
final_rice_df.head(2) 
rice_new = final_rice_df[['Area', 'rice production']].copy()
rice_new = rice_new.set_index("Area")
#rice_new
```

```python

ax = rice_new[['rice production']].plot(kind='bar', title ="Top 10 World Rice Producers", figsize=(10, 5), legend=True, fontsize=12)
plt.ylabel("Total Rice Production-1000 tons")
```
